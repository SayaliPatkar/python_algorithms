## Find the Fibonacci number of a given input and then show a graph of the programâ€™s runtime. 
1. Simple brute force recursion 
2. Dynamic programming

```python
import time
import numpy as np
import matplotlib.pyplot as plt
```

```python
def simple_plot(x, y, labels):
    # plotting the points  
    plt.plot(x, y) 
    for x_, y_, label in zip(x, y, labels):
        plt.annotate(label, # this is the text
                 (x_,y_), # these are the coordinates to position the label
                 ha='center',
                fontsize=6) # horizontal alignment can be left, right or center
    # naming the x axis 
    plt.xlabel('fibo digits') 
    # naming the y axis 
    plt.ylabel('runtime millisec') 

    # giving a title to my graph 
    plt.title('Time Plot') 

    # function to show the plot 
    plt.show()
```

```python
def brute_force_fibonacci(num: int) ->int:
    if num == 0:
        return 0
    elif num == 1:
        return 1
    else:
        return brute_force_fibonacci(num-1) + brute_force_fibonacci(num-2)
    
```

```python
calculated_fibo = {}
def dynamic_fibonacci(num: int) ->int:
    if num == 0:
        return 0
    elif num == 1:
        return 1
    elif num in calculated_fibo:
        return calculated_fibo[num]
    else:
        calculated_fibo[num] = dynamic_fibonacci(num-1) + dynamic_fibonacci(num-2)
        return calculated_fibo[num]
    
```

```python
numbers = 30
x = [num for num in range(0, numbers+1, 1)]
times_list = []
labels = []
for digit in x:
    t0 = time.time_ns()
    fibo = brute_force_fibonacci(digit)
    labels.append(f"f({digit}) = {fibo}")
    times_list.append(time.time_ns() - t0)
print(labels)
simple_plot(x, times_list, labels)
```

```python
numbers = 30
x = [num for num in range(0, numbers+1, 1)]
times_list = []
labels = []
for digit in x:
    t0 = time.time_ns()
    fibo = dynamic_fibonacci(digit)
    labels.append(f"{fibo}")
    times_list.append(time.time_ns() - t0)
print(labels)
simple_plot(x, times_list, labels)
```
